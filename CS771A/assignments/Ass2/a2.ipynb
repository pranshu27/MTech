Q3.To derive MLE  for the parameters of a k-sided multinomial distribution:
<dr>
    
<dr> A k-sided Multinomial distribution arises when a sequence of n independent and identical experiment is performed and each  experiment can result any one of k possible outcomes with respective probabilities say $p_1,p_2,...,p_r$ such that           $\sum_{i=1}^k p_i =1$.  If we let $X_i$ denote the number of the n experiments that results in outcome number i, then
    
 <dr> $P(X_1=x_1, X_2=x_2,...,X_k=x_k)={\frac{n!}{{x_1!}{x_2!}...{x_k!}}} {{p_1^{x_1}}{p_2^{x_2}}...{p_k^{x_k}}}$ 
     
 <dr> whenever $\sum_{i=1}^k x_i =n $
     
<dr> So the likelihood function of the sample denote by $L(x_1,x_2,...,x_k;p_1,p_2,...,p_k)=P(X_1=x_1, X_2=x_2,...,X_k=x_k)$
    =${\frac{n!}{{x_1!}{x_2!}...{x_k!}}} {{p_1^{x_1}}{p_2^{x_2}}...{p_k^{x_k}}}$=$L(p)$ ,say.
    
<dr>
    
<dr> 
 So, $ l(p)=\log L(p)=\log n! + \sum_{i=1}^k x_i\log{p_i} -\sum_{i=1}^k \log {x_i!}$
    
    
 <dr> Now Posing a constraint $\sum_{i=1}^k p_i =1$ with Lagrange multiplier we get:
     
  <dr> $l(p,\lambda) =l(p) + {\lambda}{ (1 - \sum_{i=1}^k p_i)}$
      
 <dr>
     
 Now to find maximum of $L(p,\lambda)$ :
     
  <dr> $\frac{\partial l(p,\lambda)}{\partial p_1} = \frac{\partial l(p)}{\partial p_1}+ \frac{\partial ({\lambda}{ (1 - \sum_{i=1}^k p_i)})}{\partial p_1}$ =$\frac{\partial{(\sum_{i=1}^k x_i \log{p_i})}}{\partial p_1}$ -${\lambda}
{\frac{\partial {( \sum_{i=1}^k p_i)}}{\partial p_1}}$= $\frac{x_1}{p_1} -\lambda$

 <dr>Similarly
     
  <dr> 
 
 $\frac{\partial l(p,\lambda)}{\partial p_2} = \frac{\partial l(p)}{\partial p_2}+ \frac{\partial ({\lambda}{ (1 - \sum_{i=1}^k p_i)})}{\partial p_2}$ =$\frac{\partial{(\sum_{i=1}^k x_i \log{p_i})}}{\partial 2}$ -${\lambda}
{\frac{\partial { (\sum_{i=1}^k p_i)})}{\partial p_2}}$= $\frac{x_1}{p_2} -\lambda$
   
  <dr>......
      
  <dr>......
      
  <dr>$\frac{\partial l(p,\lambda)}{\partial p_k} = \frac{\partial l(p)}{\partial p_k}+ \frac{\partial ({\lambda}{ (1 - \sum_{i=1}^k p_i)})}{\partial p_k}$ =$\frac{\partial{(\sum_{i=1}^k x_i \log{p_i})}}{\partial p_k}$ -${\lambda}
{\frac{\partial { (\sum_{i=1}^k p_i)}}{\partial p_k}}$= $\frac{x_1}{p_k} -\lambda$
      
 <dr>The likelihood equations are given by :
     
  <dr>$\frac{\partial l(p,\lambda)}{\partial p_1}=\frac{x_1}{p_1} -\lambda =0$
      
   <dr>$\frac{\partial l(p,\lambda)}{\partial p_2}=\frac{x_1}{p_2} -\lambda =0$
       
   <dr>......
              
   <dr> $\frac{\partial l(p,\lambda)}{\partial p_1}=\frac{x_1}{p_1} -\lambda =0$
   
  <dr> Solving which we get :
      
   <dr> $ p_1= \frac{x_1}{\lambda},p_2= \frac{x_2}{\lambda}...p_k= \frac{x_k}{\lambda} $
       
<dr> 
     Now , we got $p_i =\frac{x_i}{\lambda} $
     
<dr> 
    $\implies \sum_{i=1}^k p_i = \sum_{i=1}^k{\frac{x_i}{\lambda}}$
    
<dr>
    
   $\implies 1= {\frac{1}{\lambda}}{\sum_{i=1}^k x_i}$
    
   <dr> $\implies \lambda =n$
       
   <dr> Thus  $p_1= \frac{x_1}{n},p_2= \frac{x_2}{n}...p_k= \frac{x_k}{n}$
       
       
  <dr> Lets calculate the the kxk hessian  now:
      
   <dr> 
       Hessian Matrix = $\begin{pmatrix} \frac{\partial^2 l(p,\lambda)}{\partial^2 p_1} &\frac{\partial^2 l(p,\lambda)}{\partial p_1\partial p_2} &\cdots &\frac{\partial^2 l(p,\lambda)}{\partial p_1 \partial p_2}\\\frac{\partial^2 l(p,\lambda)}{\partial p_2\partial p_1} &\frac{\partial^2 l(p,\lambda)}{\partial^2 p_2}& \cdots &\frac{\partial^2 l(p,\lambda)}{\partial p_2 \partial p_k} \\\vdots &\vdots & \ddots &\vdots \\\frac{\partial^2 l(p,\lambda)}{\partial p_k \partial p_1}&\frac{\partial^2 l(p,\lambda)}{\partial p_k \partial p_2} & \cdots  & \frac{\partial^2 l(p,\lambda)}{\partial^2 p_k} \end{pmatrix}    
 = \begin{pmatrix} {-}{\frac{x_1}{{p_1}^2}}& 0 & \cdots & 0 \\ 0 & {-}{\frac{x_1}{{p_2}^2}} &\cdots & 0\\\vdots &\vdots & \ddots &\vdots \\ 0 & 0 & \cdots & {-}{\frac{x_1}{{p_1}^k}} \end{pmatrix}$
           
 <dr>
     
  Which is negative, as the diagonals are negative. Therefore the likelihood function  is maximum.
     
 <dr>
     
  $\therefore (p_1= \frac{x_1}{n},p_2= \frac{x_2}{n}...p_k= \frac{x_k}{n})$ is the Maximum lilelihood estimate of k-th fold multinomial distribution.
